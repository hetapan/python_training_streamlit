```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [100,105,108]
y = [10,20,30]

flg = plt.figure(figsize=(5,5))
ax = plt.axes()

ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()

ax.plot(x,y, marker="o", color="red", linestyle="--")
plt.show()
```

```python
x = np.random.randint(1,101,200)
y = np.random.randint(1,101,200)

flg = plt.figure(figsize=(5,5))
ax = plt.axes()

ax.scatter(x, y)
plt.show()
```

```python
flg = plt.figure()

ax1 = flg.add_subplot(221)
ax2 = flg.add_subplot(222)
ax3 = flg.add_subplot(223)
ax4 = flg.add_subplot(224)

plt.subplots_adjust(wspace=0.5, hspace=0.5)

ax1.set_title('ax1')
x = [105,110,120,125]
y = [10,11,12,14]
ax1.plot(x,y)

ax2.set_title('ax2')
x = ['a','b','c','d']
y = [10,11,12,14]
ax2.bar(x,y)

plt.show()
```
