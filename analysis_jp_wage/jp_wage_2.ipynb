```python
### データのimportと整形
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
# ファイルの読み込み
df_jp_ind = pd.read_csv('./csv_data/雇用_医療福祉_一人当たり賃金_全国_全産業.csv', encoding='shift_jis')
df_jp_category = pd.read_csv('./csv_data/雇用_医療福祉_一人当たり賃金_全国_大分類.csv', encoding='shift_jis')
df_pref_ind = pd.read_csv('./csv_data/雇用_医療福祉_一人当たり賃金_都道府県_全産業.csv', encoding='shift_jis')
df_pref_category = pd.read_csv('./csv_data/雇用_医療福祉_一人当たり賃金_都道府県_大分類.csv', encoding='shift_jis')
```

```python
df_jp_ind[df_jp_ind['年齢'] == '年齢計']
```

```python
df_diff = df_jp_ind[df_jp_ind['年齢'] == '年齢計']
```

```python
# 1行前の差分を取る
df_diff['一人当たり賃金の差分（万円）'] = df_diff['一人当たり賃金（万円）'].diff()
df_diff
```

```python
# グルーピングと平均の算出
df_jp_ind.groupby('年齢')['一人当たり賃金（万円）'].mean()
```

```python
df_jp_ind.groupby('年齢')['所定内給与額（万円）','一人当たり賃金（万円）', '年間賞与その他特別給与額（万円）'].mean()
```

```python
df_temp = df_jp_category[df_jp_category['年齢'] == '年齢計']
df_temp.groupby('産業大分類名')['一人当たり賃金（万円）'].mean()
```

```python
df_temp_2 = df_pref_ind[df_pref_ind['年齢'] == '年齢計']
df_temp_2.groupby('都道府県名')['所定内給与額（万円）','一人当たり賃金（万円）', '年間賞与その他特別給与額（万円）'].mean()
```

```python
df_temp_2.loc[df_temp_2['一人当たり賃金（万円）'].idxmax()]
```

```python
df_pref_category[df_pref_category['所定内給与額（万円）'] == '-']

df_pref_category['所定内給与額（万円）'] = df_pref_category['所定内給与額（万円）'].replace('-', np.nan)
df_pref_category['一人当たり賃金（万円）'] = df_pref_category['一人当たり賃金（万円）'].replace('-', np.nan)
df_pref_category['年間賞与その他特別給与額（万円）'] = df_pref_category['年間賞与その他特別給与額（万円）'].replace('-', np.nan)
df_pref_category.dropna(subset=['所定内給与額（万円）'], axis=0, inplace=True)
df_pref_category = df_pref_category.astype({'所定内給与額（万円）': float, '一人当たり賃金（万円）': float, '年間賞与その他特別給与額（万円）': float})

df_temp_3 = df_pref_category[df_pref_category['年齢'] == '年齢計']
df_temp_3_gr = df_temp_3.groupby(['都道府県名','産業大分類名'])['所定内給与額（万円）','一人当たり賃金（万円）', '年間賞与その他特別給与額（万円）'].mean()
df_temp_3_gr.loc['三重県':'京都府']
```
